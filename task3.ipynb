```python
import numpy as np
#y_actual and y_pred are arrays representing actual targets and predicting targets respectively.
#arrays can have single element


def mse(y_actual, y_pred):
  sq= np.square(y_actual - y_pred)
  return np.mean(sq)
```

image.png

```python
def rmse(y_actual, y_pred):
  return np.sqrt(mse(y_actual, y_pred))
```

**Cosine similarity** is a mathematical metric used to measure the similarity between two vectors in a multi-dimensional space, particularly in high-dimensional spaces, by calculating the cosine of the angle between them

image.png

```python
def cosine_similarity(v1, v2):

  dot_product = np.dot(v1, v2)
  magnitude_v1 = np.linalg.norm(v1) #magnitude of the vector1
  magnitude_v2 = np.linalg.norm(v2) #magnitude of the vector2
  if magnitude_v1 == 0 or magnitude_v2 == 0:
      return 0  # Handle cases where one or both vectors have zero magnitude
  return dot_product / (magnitude_v1 * magnitude_v2)
```

```python
y_t = np.array([1, 2, 3, 4, 5])
y_p = np.array([1.2, 1.8, 3.1, 3.9, 5.2])
print("MSE:", mse(y_t, y_p))
print("RMSE:", rmse(y_t,y_p))

v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
print("Cosine Similarity:", cosine_similarity(v1, v2))
```

```python
def softmax(x):
  exp_x = np.exp(x - np.max(x))  #Applies the exponential function (e^x) to each element in x. Subtract max for numerical stability
  return exp_x / exp_x.sum(axis=0) #Divides each exponentiated value by the sum of all exponentiated value.
```

```python
x = np.array([1,2,3])
print("Softmax:", softmax(x))
```

```python
def linear_regression_gradient_descent(X, y, learning_rate=0.01, epochs=1000):
  #epochs are the number of iterations
  m, n = X.shape  # Number of samples (m) and features (n)
  weights = np.zeros(n)
  intercept = 0

  for _ in range(epochs):
        # predictions
        y_pred = np.dot(X, weights) + intercept

        dw = (1 / m) * np.dot(X.T, (y_pred - y))  # Gradient/slope for weights
        db = (1 / m) * np.sum(y_pred - y)         # Gradient for intercept

        # Update weights and intercept
        weights -= learning_rate * dw
        intercept -= learning_rate * db

  return weights, intercept
```

```python
X = np.array([[1], [2], [3], [4]])  # Feature matrix (single feature)
y = np.array([2, 4, 6, 8])  # Target values

# model
weights, intercept = linear_regression_gradient_descent(X, y)

print("Weights:", weights)
print("Intercept:", intercept)
```
